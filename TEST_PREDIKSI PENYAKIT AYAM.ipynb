```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Muat model
model = load_model('modelxception (1).h5')  # Nama file model Anda

# Ganti nama file sesuai gambar yang diupload
img_path = '17.jpg'  # Gambar input

# Tampilkan gambar
img_display = image.load_img(img_path)
plt.imshow(img_display)
plt.axis('off')
plt.title("Gambar yang Diprediksi")
plt.show()

# Preprocessing untuk model
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = x / 255.0  # Normalisasi

# Prediksi
prediction = model.predict(x)

# Label kelas (urutan harus sesuai saat training)
class_labels = ['Coccidiosis', 'Healthy', 'New Castle Disease', 'Salmonella']

# Ambil hasil prediksi tertinggi
predicted_index = np.argmax(prediction)
predicted_label = class_labels[predicted_index]
confidence = prediction[0][predicted_index] * 100

print(f"Hasil Prediksi: {predicted_label} ({confidence:.2f}%)")
```
